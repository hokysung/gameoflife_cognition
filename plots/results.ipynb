```python
fig = plt.figure(figsize=(20,30))

plt.title("Loss with Ordered Dataset", fontsize=30)
for iteration in range(10):
    ax = fig.add_subplot(5,2,iteration+1)
    path = '../exp_stossp_ep50_lr001_ordered_baseline/run_{}/'.format(iteration)
    losses = []
    for i in range(4):
        losses.append(np.load(path+"loss_{}.npy".format(i)))
    test_loss, mcc = np.load(path+"test_res.npy")
    losses = np.concatenate([losses[i] for i in range(4)])
    ax.plot(range(len(losses)), losses[:,0], label="Train Loss")
    ax.plot(range(len(losses)), losses[:,1], label="Validation Loss")
    ax.set_title('MCC: {}'.format(mcc))
    ax.set_xlabel('training epochs')
    ax.set_ylabel('loss')
    ax.legend()
    
plt.show()
```
